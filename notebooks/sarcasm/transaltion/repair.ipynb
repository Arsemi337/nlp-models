```python
import pandas as pd
data_df = pd.read_json('../../../data/translated/sarcasm/sarcasm_headlines_dataset_pl6.json')
data_df
```

```python
translated_list = data_df["headline_pl"].values.tolist()
translated_list
```

```python
import json

def parse_sarcasm_json_dataset(file):
    for line in open(file, 'r'):
        yield json.loads(line)
```

```python
data = list(parse_sarcasm_json_dataset('../../../data/news_headlines_for_sarcasm_detection/Sarcasm_Headlines_Dataset_v2.json'))

df = pd.DataFrame(data)
# df[10000:12000]
```

```python
new_df = df[12000:15000].copy(deep=True)
new_df
```

```python
new_df.reset_index(drop=True, inplace=True)
new_df
```

```python
new_df['headline_pl'] = translated_list
new_df
```

```python
new_df.to_json(r'../../../data/translated/sarcasm/good_12000_15000.json')
```
